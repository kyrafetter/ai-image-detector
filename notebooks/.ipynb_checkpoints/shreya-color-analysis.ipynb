<h4> EDA Step 3: Plotting Image Brightness Across AI-Generated and Real Images </h4>

```python
ai_generated_brightness = []
real_brightness = []

for img_path, label in zip(images, labels): # iterate through image and class label
    # loading each image in grayscale mode to simplify brightness calculation by reducing to a single channel (0 to 255 pixel values)
    # grayscale represents intensity/brightness better because it gets the weighted average of the RGB color channels
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    # average brightness: calculate the mean pixel intensity for the grayscale image, which represents the image's overall brightness
    average_brightness = np.mean(img)

    if label == 0:  # label 0 is for ai-generated images
        ai_generated_brightness.append(average_brightness)
    else:           # label 1 is for real images
        real_brightness.append(average_brightness)
```

```python
avg_real_brightness = np.mean(real_brightness)
avg_ai_generated_brightness = np.mean(ai_generated_brightness)

plt.figure(figsize=(6, 4))
plt.bar(['Real Images', 'AI-Generated Images'], [avg_real_brightness, avg_ai_generated_brightness], color=['blue', 'orange'])
plt.ylabel("Average Brightness")
plt.title("Average Brightness of Real vs. AI-Generated Images")
plt.show()
```

```python
# Generating indices for each image type
ai_generated_indices = list(range(len(ai_generated_brightness)))
real_indices = list(range(len(real_brightness)))

# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(ai_generated_indices, ai_generated_brightness, color='orange', alpha=0.5, label='AI-Generated Images')
plt.scatter(real_indices, real_brightness, color='blue', alpha=0.5, label='Real Images')
plt.xlabel("Image Index")
plt.ylabel("Brightness")
plt.title("Brightness Comparison between AI-Generated and Real Images")
plt.legend()
plt.show()
```

<h3> Analysis Plotting Image Brightness Across AI-Generated and Real Images:  </h3>

Based on the bar plot and scatterplot, it seems that the real images are generally brighter than the ai-generated images. 
This brightness variation is important because we can choose CNN layers that focus on brightness gradients helping distinguish AI-generated images and real images better. We could also perform pre-processing to standardize brightness across images before the images enter the CNN, so that the model doesn't make predictions purely based off of brightness. 

<h4> EDA Step 4: Plotting Aspect Ratios Across All Images </h4>

```python
# dimensions is (width, height) tuples for each image
dimensions = []

for img_path in images:
    img = cv2.imread(img_path)
    height, width = img.shape[:2]
    dimensions.append((width, height))  # Append as (height, width, # color channels) <-- we only grab the first 2


aspect_ratios = [width / height for width, height in dimensions]
plt.figure(figsize=(10, 6))
plt.hist(aspect_ratios, bins=20, color='skyblue', edgecolor='black')
plt.xlabel("Aspect Ratio (Width / Height)")
plt.ylabel("Frequency")
plt.title("Distribution of Aspect Ratios in the Dataset")
plt.show()
```

<h4> Aspect Ratio Analysis: </h4>

It seems that not all images are square or have an aspect ratio (width/height) of 1 as there are several distinct peaks in the histogram. Therefore, it is possible that cropping all images to 128 x 128 pixels could cause the images to loose important visual information required for the CNN layers to extract features. We may need to resize by adding padding for non square images to fit them into a square frame withotu losing aspect ratio integrity.
