```python
import matplotlib.pyplot as plt
from PIL import Image
import os
import sys

import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import random
import cv2
import numpy as np

```

```python
# Note - add all images to this
# Define ImageDataGenerator for training
# Currently not performing train/test split, just using all images to perform EDA
datagen = ImageDataGenerator(
)

# Loading data
data = datagen.flow_from_directory(
    'c:/Users/annet/OneDrive/Desktop/CSE 151A project/ai-image-detector/data/images',
    class_mode='binary',   # Performing binary classification
)

images = data.filepaths
labels = data.classes
```

- cv2.Laplacian(img, cv2.CV_64F) computes the Laplacian of the image, emphasizing areas with rapid intensity changes.
- .var() calculates the variance of these Laplacian values, which serves as an estimate of image sharpness.


Higher variance → sharper image (more detail and edges).

Lower variance → blurrier image.

```python
ai_generated_sharpness = []
real_img_sharpness = []

for img_path, label in zip(images, labels):  # iterate through image and class label
    # Load the image in grayscale mode
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    
    # Calculate the Laplacian (second derivative) of the image
    laplacian_var = cv2.Laplacian(img, cv2.CV_64F).var()
    
    # Append to the appropriate list based on the label
    if label == 0:  # label 0 for ai-generated images
        ai_generated_sharpness.append(laplacian_var)
    else:           # label 1 for real images
        real_img_sharpness.append(laplacian_var)
```

```python
avg_real_img_sharpness = np.mean(real_img_sharpness)
avg_ai_generated_brightness = np.mean(ai_generated_sharpness)
```

### Plotting image sharpness for both AI generated images and Real images

```python

# Plotting data points with mean lines for each category
plt.figure(figsize=(10, 6))

# Plot individual sharpness points
plt.plot(ai_generated_sharpness, label="AI Generated Sharpness", marker='o', linestyle='', markersize=6)
plt.plot(real_img_sharpness, label="Real Image Sharpness", marker='x', linestyle='', markersize=6)

# Adding mean lines for each category
plt.axhline(y=avg_ai_generated_sharpness, color='blue', linestyle='--', label="Avg AI Generated Sharpness")
plt.axhline(y=avg_real_img_sharpness, color='orange', linestyle='--', label="Avg Real Image Sharpness")

# Labels and Title
plt.xlabel("Image Index")
plt.ylabel("Sharpness (Variance of Laplacian)")
plt.title("Sharpness Comparison between AI Generated and Real Images with Averages")
plt.legend()

# Display the graph
plt.show()
```

Based on the graph, AI generated images tend to have sharper images than the real images. This maybe inconclusive because it seems that it is random and the sample I am using is very small.
