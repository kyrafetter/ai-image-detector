```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models
from PIL import Image
import matplotlib.pyplot as plt
import random
import cv2
import numpy as np
from collections import Counter
import os
import sys
import glob
import pandas as pd
```

### Data Preprocessing
- rezizing the images <br> 
- normalizing the images  <br>
- splitting the data for training and testing <br>

```python
# dir: Directory contains "real" and "fake" directories of images
dir = "../../data/images"

# datagen: setting up ImageDataGenerator for augmentation and data preprocessing
datagen = ImageDataGenerator(
    rescale=1.0/255.0,      # Normalizing the pixels to be in [0,1] range by dividing by 255.0
    validation_split=0.2    # data validation: 20%
)

# Training data setup
train_data = datagen.flow_from_directory(
    dir,
    target_size=(128,128),  # 1 - resize images to 128x128
    batch_size=32,          # 2 - batch setup - number of images per batch
    class_mode='binary',    # 3 - binary classification - 2 classes (fake and real)
    subset='training'
)

# Validation data setup
validation_data = datagen.flow_from_directory(
    dir,
    target_size=(128,128),  # 1 - resize images to 128x128
    batch_size=32,          # 2 - batch setup - number of images per batch
    class_mode='binary',    # 3 - binary classification - 2 classes (fake and real)
    subset='validation'
)

# Accessing images and labels for checking setup
images = train_data.filepaths
labels = train_data.classes

real_images = [images[i] for i in range(len(images)) if labels[i] == 0]
fake_images = [images[i] for i in range(len(images)) if labels[i] == 1]
```

### Setting up the CNN Model 
- defined basic CNN model for binary image classification using Tensorflow and Keras API 

```python
# imported above: from tensorflow.keras import layers, models
# layers - module that provides the building blocks for the neural network (convolutional, pooling, and dense (fully connected) layers)
# models - module to create structure of neural network

# Defining the model
# Sequential model defined:
#    - we have exactly one input tensor (image) and one output tensor (binary classification)
#    - linear stack of layers where layers are arranged in seuqence (feed-forward architecture)
#    - we don't have any branching or skipping, so sequential is suitable for each layer's output to feed directly into the next layer
model_thirty_images = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(128,128,3)),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(128, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(1, activation='sigmoid')                                       # Sigmoid - being used for binary classification
])

model_thirty_images.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

### Training the Model

```python
# Training the model
history = model_thirty_images.fit(
    train_data,
    validation_data=validation_data,
    epochs=10  # Number of epochs
)
```

### Evaluating the Model 

```python
# Evaluating the model
loss, accuracy = model_thirty_images.evaluate(validation_data)
# Validation loss = error/difference between actual labels and the model's predictions on validation set
#   - validation loss calculated on loss function - binary cross entropy since we have binary classification
print(f"Validation Loss: {loss}")
# Validation accuracy = percentage of correct predictions model makes on validation set
print(f"Validation Accuracy: {accuracy}")
```

### Plotting Training Results

```python
# Plotting the training and validation accuracy results
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
# Plotting the training and validation loss results
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
