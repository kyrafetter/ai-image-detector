### Imports

```python
import matplotlib.pyplot as plt
from PIL import Image
import os
import sys
import glob
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

### Load Images

```python
datagen = ImageDataGenerator(
)

# Loading data
data = datagen.flow_from_directory(
    '../../../../data',

    class_mode='binary',   # Performing binary classification
)

images = data.filepaths
labels = data.classes
```

```python
real_images = [images[i] for i in range(len(images)) if labels[i] == 0]
fake_images = [images[i] for i in range(len(images)) if labels[i] == 1]
```

### Plot Density Plot (All Pixels)

```python
def calculate_color_densities(image_paths):
    density_r = np.zeros(256)
    density_g = np.zeros(256)
    density_b = np.zeros(256)
    total_pixels = 0
    
    for path in image_paths:
        image = cv2.imread(path)
        if image is None:
            print(path)
            continue
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        total_pixels += image_rgb.shape[0] * image_rgb.shape[1]
        
        density_r += cv2.calcHist([image_rgb], [0], None, [256], [0, 256]).flatten()
        density_g += cv2.calcHist([image_rgb], [1], None, [256], [0, 256]).flatten()
        density_b += cv2.calcHist([image_rgb], [2], None, [256], [0, 256]).flatten()
    
    # Normalize to convert to density
    density_r /= total_pixels
    density_g /= total_pixels
    density_b /= total_pixels
    
    return density_r, density_g, density_b
```

```python
real_density_r, real_density_g, real_density_b = calculate_color_densities(real_images)
fake_density_r, fake_density_g, fake_density_b = calculate_color_densities(fake_images)
```

```python
# Plot the density histograms
plt.figure(figsize=(13, 3))

colors = {1 : "Red", 2 : "Green", 3: "Blue"}

for i in range(1, 4):
    plt.subplot(1, 3, i)
    color = colors[i]
    if color == "Red":
        plt.plot(real_density_r, color='red', label='Real')
        plt.plot(fake_density_r, color='darkred', label='Fake', linestyle='--')
    elif color == "Green":
        plt.plot(real_density_g, color='green', label='Real')
        plt.plot(fake_density_g, color='darkgreen', label='Fake', linestyle='--')
    else:
        plt.plot(real_density_b, color='blue', label='Real')
        plt.plot(fake_density_b, color='darkblue', label='Fake', linestyle='--')
    plt.title(color + ' Channel Density')
    plt.xlabel('Pixel Intensity')
    plt.ylabel('Density')
    plt.legend()

plt.tight_layout()
plt.show()
```

### Plot Mean Color Value

```python
def calculate_mean_rgb(image_paths):
    mean_r = []
    mean_g = []
    mean_b = []
    
    for path in image_paths:
        image = cv2.imread(path)
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        
        mean_r.append(np.mean(image_rgb[:, :, 0]))
        mean_g.append(np.mean(image_rgb[:, :, 1]))
        mean_b.append(np.mean(image_rgb[:, :, 2]))
    
    return mean_r, mean_g, mean_b
```

```python
real_mean_r, real_mean_g, real_mean_b = calculate_mean_rgb(real_images)
fake_mean_r, fake_mean_g, fake_mean_b = calculate_mean_rgb(fake_images)
```

```python
# Plot the density histograms
plt.figure(figsize=(13, 3))

colors = {1 : "Red", 2 : "Green", 3: "Blue"}

for i in range(1, 4):
    plt.subplot(1, 3, i)
    color = colors[i]
    if color == "Red":
        plt.plot(np.sort(real_mean_r), color='red', label='Real')
        plt.plot(np.sort(fake_mean_r), color='darkred', label='Fake', linestyle='--')
    elif color == "Green":
        plt.plot(np.sort(real_mean_g), color='green', label='Real')
        plt.plot(np.sort(fake_mean_g), color='darkgreen', label='Fake', linestyle='--')
    else:
        plt.plot(np.sort(real_mean_b), color='blue', label='Real')
        plt.plot(np.sort(fake_mean_b), color='darkblue', label='Fake', linestyle='--')
    plt.title("Mean " + color + ' Channel Density')
    plt.xlabel('Image')
    plt.ylabel('Pixel Intensity')
    plt.legend()

plt.tight_layout()
plt.show()
```

### Plot Median Pixel Value

```python
def calculate_median_rgb(image_paths):
    mean_r = []
    mean_g = []
    mean_b = []
    
    for path in image_paths:
        image = cv2.imread(path)
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        
        mean_r.append(np.median(image_rgb[:, :, 0]))
        mean_g.append(np.median(image_rgb[:, :, 1]))
        mean_b.append(np.median(image_rgb[:, :, 2]))
    
    return mean_r, mean_g, mean_b
```

```python
real_median_r, real_median_g, real_median_b = calculate_median_rgb(real_images)
fake_median_r, fake_median_g, fake_median_b = calculate_median_rgb(fake_images)
```

```python
# Plot the density histograms
plt.figure(figsize=(13, 3))

colors = {1 : "Red", 2 : "Green", 3: "Blue"}

for i in range(1, 4):
    plt.subplot(1, 3, i)
    color = colors[i]
    if color == "Red":
        plt.plot(np.sort(real_median_r), color='red', label='Real')
        plt.plot(np.sort(fake_median_r), color='darkred', label='Fake', linestyle='--')
    elif color == "Green":
        plt.plot(np.sort(real_median_g), color='green', label='Real')
        plt.plot(np.sort(fake_median_g), color='darkgreen', label='Fake', linestyle='--')
    else:
        plt.plot(np.sort(real_median_b), color='blue', label='Real')
        plt.plot(np.sort(fake_median_b), color='darkblue', label='Fake', linestyle='--')
    plt.title("Median " + color + ' Channel Density')
    plt.xlabel('Image')
    plt.ylabel('Pixel Intensity')
    plt.legend()

plt.tight_layout()
plt.show()
```
