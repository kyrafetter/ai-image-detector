```python
import matplotlib.pyplot as plt
from PIL import Image
import os
import sys

import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import random
import cv2
import numpy as np

import pandas as pd
```

```python
# Note - add all images to this
# Define ImageDataGenerator for training
# Currently not performing train/test split, just using all images to perform EDA
datagen = ImageDataGenerator(
)

# Loading data
data = datagen.flow_from_directory(
    'c:/Users/annet/OneDrive/Desktop/CSE 151A project/ai-image-detector/data/images',
    class_mode='binary',   # Performing binary classification
)

images = data.filepaths
labels = data.classes
```

```python
ai_generated_sharpness = []
real_img_sharpness = []

for img_path, label in zip(images, labels):
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    if img is None:
        continue
    laplacian_var = cv2.Laplacian(img, cv2.CV_64F).var()
    if label == 0:
        ai_generated_sharpness.append(laplacian_var)
    else:
        real_img_sharpness.append(laplacian_var)
```

```python
real_img_data = pd.DataFrame(real_img_sharpness)
ai_generated_data = pd.DataFrame(ai_generated_sharpness)
print("real_img_data: ")
print(real_img_data.describe())

print("ai_generated_data: ")
print(ai_generated_data.describe())

avg_real_img_sharpness = np.mean(real_img_sharpness)
avg_ai_generated_sharpness = np.mean(ai_generated_sharpness)

```

### Plotting image sharpness for both AI generated images and Real images

```python

# Plotting data points with mean lines for each category
plt.figure(figsize=(10, 6))

# Plot individual sharpness points
plt.plot(ai_generated_sharpness, label="AI Generated Sharpness", marker='o', linestyle='', markersize=6)
plt.plot(real_img_sharpness, label="Real Image Sharpness", marker='x', linestyle='', markersize=6)

# Adding mean lines for each category
plt.axhline(y=avg_ai_generated_sharpness, color='blue', linestyle='--', label="Avg AI Generated Sharpness")
plt.axhline(y=avg_real_img_sharpness, color='orange', linestyle='--', label="Avg Real Image Sharpness")

# Labels and Title
plt.xlabel("Image Index")
plt.ylabel("Sharpness (Variance of Laplacian)")
plt.title("Sharpness Comparison between AI Generated and Real Images with Averages")
plt.legend()

# Display the graph
plt.show()
```

### Analysis
1. Sharper Images: AI-generated images tend to have higher sharpness on average (indicated by a higher mean sharpness). This could imply that AI models are tuned to produce images that appear more defined or crisp.

2. Less Variation in AI Sharpness: The lower standard deviation in AI-generated images implies that the sharpness is more consistent across the dataset, with fewer variations in sharpness levels compared to real images.

3. Higher Median Sharpness: The higher median in AI-generated images suggests that even the "middle" sharpness values in AI images are higher than those in real images, reinforcing the consistency in sharpness across AI images.

This analysis suggests AI images are optimized for sharper, more consistent quality, whereas real images display a broader range of sharpness, likely due to diverse conditions in real-life photography (lighting, focus, etc.).
