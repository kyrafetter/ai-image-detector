##### Show image padding function

```python
##Example of pad_to_square transform seen below

def pad_to_square(image):
    width, height = image.size
    max_dim = max(width, height)
    padding = (
        (max_dim - width) // 2,  # left padding
        (max_dim - height) // 2,  # top padding
        (max_dim - width + 1) // 2,  # right padding (handle odd numbers)
        (max_dim - height + 1) // 2  # bottom padding (handle odd numbers)
    )
    return transforms.functional.pad(image, padding, fill=0)

input_image_path = os.path.join(dir, "real/0015.jpg")
image = Image.open(input_image_path)


plt.imshow(image)
plt.axis('off')  # Hide the axes for better visualization
plt.text(
    10, 10, "Original image",
    color='red', fontsize=14,
    bbox=dict(facecolor='white', alpha=0.7)
)
plt.show()


# Apply the padding function
padded_image = pad_to_square(image)
print(padded_image.size)

plt.imshow(padded_image)
plt.axis('off')  # Hide the axes for better visualization
plt.text(
    10, 10, "Padded image",
    color='red', fontsize=14,
    bbox=dict(facecolor='white', alpha=0.7)
)
plt.show()
```

##### Defined image pad function and split data

```python
#Define transforms/ functions
class pad_to_square(object):        #Defined as a class for model compatibility
    def __call__(self,image):
        width, height = image.size
        max_dim = max(width, height)
        padding = (
            (max_dim - width) // 2,  # left padding
            (max_dim - height) // 2,  # top padding
            (max_dim - width + 1) // 2,  # right padding (handle odd numbers)
            (max_dim - height + 1) // 2  # bottom padding (handle odd numbers)
        )
        return transforms.functional.pad(image, padding, fill=0)
    def __repr__(self):
        return self.__class__.__name__+'()'


def save_checkpoint(state, is_best, checkpoint_dir):
    filepath = os.path.join(checkpoint_dir, 'checkpoint.pth')
    torch.save(state, filepath)
    if is_best:
        shutil.copyfile(filepath, os.path.join(checkpoint_dir, 'model_best.pth.tar'))
```

### Setting up the CNN Model 
- defined basic CNN model for binary image classification using Tensorflow and Keras API 

```python
class CNNModel(nn.Module):
    def __init__(self):
        super(CNNModel, self).__init__()
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3, padding=1)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        self.conv3 = nn.Conv2d(64, 128, kernel_size=3, padding=1)
        self.flatten = nn.Flatten()
        self.fc1 = nn.Linear(128 * 16 * 16, 128)  # Adjust the size based on your input
        self.fc2 = nn.Linear(128, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.pool(nn.ReLU()(self.conv1(x)))
        x = self.pool(nn.ReLU()(self.conv2(x)))
        x = self.pool(nn.ReLU()(self.conv3(x)))
        x = self.flatten(x)
        x = nn.ReLU()(self.fc1(x))
        x = self.sigmoid(self.fc2(x))
        return x
```

### Setting Up the Loss Function and Optimizer

```python
model = CNNModel().to(device)   #Define model and send to specified cpu or gpu
criterion = nn.BCELoss()  # Binary Cross Entropy Loss for binary classification
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

## 2: Train your first model

### Training the Model and Validation where validation metrics are used to improve and tune the model during training 

```python
num_epochs = 5
best_perf = 0
train_accuracy_list = []
val_accuracy_list = []
train_loss_list = []
val_loss_list = []


if not os.path.exists('../../result'):
    os.makedirs('../../result')

checkpoint_dir = '../../result'

if not os.path.exists(checkpoint_dir):
    os.makedirs(checkpoint_dir)

for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    correct_train = 0
    total_train = 0

    for images, labels in tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs}", leave=True):
        #Send data to specified cpu or gpu device
        images = images.to(device)
        labels = labels.to(device)

        labels = labels.unsqueeze(1).float()  # Reshape labels and convert to float
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()

        # Calculate training accuracy
        predicted_train = (outputs > 0.5).float()
        correct_train += (predicted_train == labels).sum().item()
        total_train += labels.size(0)

    # Appends average training loss for the epoch
    train_loss_list.append(running_loss / len(train_loader))
    # Appends training accuracy for the epoch
    train_accuracy_list.append(correct_train / total_train)

    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {running_loss/len(train_loader)}")

    # Validation Step
    model.eval()
    val_loss = 0.0
    correct = 0
    total = 0
    with torch.no_grad():
        for images, labels in val_loader:
            #Send data to specified cpu or gpu device
            images = images.to(device)
            labels = labels.to(device)

            labels = labels.unsqueeze(1).float()
            outputs = model(images)
            loss = criterion(outputs, labels)
            val_loss += loss.item()
            predicted = (outputs > 0.5).float()
            correct += (predicted == labels).sum().item()
            total += labels.size(0)

    # Appends average validation loss for the epoch
    val_loss_list.append(val_loss / len(val_loader))
    # Appends validation accuracy for the epoch
    val_accuracy_list.append(correct / total)
    val_accuracy = correct / total
    print(f"Validation Loss: {val_loss/len(val_loader)}, Validation Accuracy: {val_accuracy}")


    # Save model checkpoint if best accuracy
    perf = val_accuracy

    is_best = perf > best_perf
    best_perf = max(perf,best_perf)
    print('best acc: {:1f}'.format(best_perf))
    save_checkpoint({
        'epoch': epoch + 1,
        'state_dict': model.state_dict(),
        'best_perf': best_perf,
        'optimizer': optimizer.state_dict(),
    }, is_best, checkpoint_dir)
```

```python
dataset = datasets.ImageFolder(root=dir, transform=transform)

# Print class-to-index mapping
print(dataset.class_to_idx)
```

```python
from sklearn.model_selection import KFold
from torch.utils.data import Subset

# Parameters
k_folds = 5
num_epochs = 5
batch_size = 32
seed = 42
learning_rate = 0.001
torch.manual_seed(seed)

# KFold object
kfold = KFold(n_splits=k_folds, shuffle=True, random_state=seed)

# Placeholder for fold results
fold_results = {}

# K-Fold Training
for fold, (train_idx, val_idx) in enumerate(kfold.split(dataset)):
    print(f'FOLD {fold + 1}/{k_folds}')
    print('--------------------------------')

    # Subset Datasets
    train_subset = Subset(dataset, train_idx)
    val_subset = Subset(dataset, val_idx)

    # DataLoaders
    train_loader = DataLoader(train_subset, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_subset, batch_size=batch_size, shuffle=False)

    # Model, Loss, Optimizer
    model = CNNModel().to(device)
    criterion = nn.BCELoss()  # Binary Cross Entropy Loss
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)

    # Training and Validation Loop
    best_val_accuracy = 0.0
    for epoch in range(num_epochs):
        # Training
        model.train()
        running_loss = 0.0
        correct_train = 0
        total_train = 0

        for images_batch, labels in tqdm(train_loader, desc=f"Fold {fold+1}, Epoch {epoch+1}/{num_epochs}", leave=False):
            images_batch, labels = images_batch.to(device), labels.to(device).unsqueeze(1).float()
            optimizer.zero_grad()
            outputs = model(images_batch)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()
            predicted_train = (outputs > 0.5).float()
            correct_train += (predicted_train == labels).sum().item()
            total_train += labels.size(0)

        train_accuracy = correct_train / total_train
        print(f'Epoch {epoch + 1}, Train Loss: {running_loss / len(train_loader):.4f}, Train Accuracy: {train_accuracy:.4f}')

        # Validation
        model.eval()
        val_loss = 0.0
        correct_val = 0
        total_val = 0
        with torch.no_grad():
            for images_batch, labels in val_loader:
                images_batch, labels = images_batch.to(device), labels.to(device).unsqueeze(1).float()
                outputs = model(images_batch)
                loss = criterion(outputs, labels)
                val_loss += loss.item()
                predicted_val = (outputs > 0.5).float()
                correct_val += (predicted_val == labels).sum().item()
                total_val += labels.size(0)

        val_accuracy = correct_val / total_val
        print(f'Validation Loss: {val_loss / len(val_loader):.4f}, Validation Accuracy: {val_accuracy:.4f}')

        # Save Best Model
        if val_accuracy > best_val_accuracy:
            best_val_accuracy = val_accuracy
            best_model_state = model.state_dict()

    print(f'Best Validation Accuracy for Fold {fold + 1}: {best_val_accuracy:.4f}')
    fold_results[fold] = best_val_accuracy

    # Save Best Model for Each Fold
    torch.save(best_model_state, f'best_model_fold_{fold+1}.pth')

# Calculate Average Accuracy
avg_accuracy = np.mean(list(fold_results.values()))
print('--------------------------------')
print(f'Average Validation Accuracy Across {k_folds} Folds: {avg_accuracy:.4f}')
```

##  3 - Evaluate your Model and Compare Training vs. Test Error

### Test the model on Training Data - using the test metrical to report final performance of model our model on training data

```python
# Calculate Training Loss and Accuracy
criterion = nn.BCELoss()
model.train()  # Set the model to training mode

train_loss = 0.0
correct = 0
total = 0
train_loss_list = []
train_accuracy_list = []

# Collect predictions and labels during the training phase
all_train_labels = []
all_train_predictions = []

for images, labels in train_loader:
    # Send data to specified CPU or GPU device
    images = images.to(device)
    labels = labels.to(device)

    labels = labels.unsqueeze(1).float()  # Reshape labels and convert to float
    outputs = model(images)
    loss = criterion(outputs, labels)
    train_loss += loss.item()

    predicted = (outputs > 0.5).float()  # Binary prediction
    correct += (predicted == labels).sum().item()
    total += labels.size(0)

    # Store labels and predictions
    all_train_labels.extend(labels.cpu().numpy())
    all_train_predictions.extend(predicted.cpu().numpy())

train_loss_list.append(train_loss / len(train_loader))
train_accuracy_list.append(correct / total)
train_accuracy = correct / total
print(f"Training Loss: {train_loss / len(train_loader)}, Training Accuracy: {train_accuracy}")

# Getting Precision, Recall, F1-Score for Training Data
all_train_labels = np.array(all_train_labels)
all_train_predictions = np.array(all_train_predictions)

precision_train = precision_score(all_train_labels, all_train_predictions)
recall_train = recall_score(all_train_labels, all_train_predictions)
f1_train = f1_score(all_train_labels, all_train_predictions)

print(f"Precision (Training): {precision_train}")
print(f"Recall (Training): {recall_train}")
print(f"F1 Score (Training): {f1_train}")

# Getting True Positives, False Positives, True Negatives, and False Negatives for Training Data
TP_train = np.sum((all_train_predictions == 1) & (all_train_labels == 1))
FP_train = np.sum((all_train_predictions == 1) & (all_train_labels == 0))
TN_train = np.sum((all_train_predictions == 0) & (all_train_labels == 0))
FN_train = np.sum((all_train_predictions == 0) & (all_train_labels == 1))

print(f"Total Number of Training Observations: {len(all_train_predictions)}")
print(f"True Positives: {TP_train}")
print(f"False Positives: {FP_train}")
print(f"True Negatives: {TN_train}")
print(f"False Negatives: {FN_train}")
```

### Test the model on Test Data - using the test metrics ot report the final performance of our model on test data

```python
#Testing Stage - after training is complete
criterion = nn.BCELoss()
model.eval()


test_loss = 0.0
correct = 0
total = 0
test_loss_list = []
test_accuracy_list = []
# Testing phase collects predictions and corresponding labels during the testing phase
all_labels = []
all_predictions = []

with torch.no_grad():
    for images, labels in test_loader:
        #Send data to specified cpu or gpu device
        images = images.to(device)
        labels = labels.to(device)

        labels = labels.unsqueeze(1).float()
        outputs = model(images)
        loss = criterion(outputs, labels)
        test_loss += loss.item()
        predicted = (outputs > 0.5).float()
        correct += (predicted == labels).sum().item()
        total += labels.size(0)

        # Store labels and predictions
        all_labels.extend(labels.cpu().numpy())
        all_predictions.extend(predicted.cpu().numpy())

test_loss_list.append(test_loss / len(test_loader))
test_accuracy_list.append(correct / total)
test_accuracy = correct / total
print(f"Testing Loss: {test_loss/len(test_loader)}, Testing Accuracy: {test_accuracy}")


# Getting Precision, Recall, F1-Score Information
all_labels = np.array(all_labels)
all_predictions = np.array(all_predictions)

precision = precision_score(all_labels, all_predictions)
recall = recall_score(all_labels, all_predictions)
f1 = f1_score(all_labels, all_predictions)

print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")

# Getting True Positives, False Positives, True Negatives, and False Negatives
TP = np.sum((all_predictions == 1) & (all_labels == 1))
FP = np.sum((all_predictions == 1) & (all_labels == 0))
TN = np.sum((all_predictions == 0) & (all_labels == 0))
FN = np.sum((all_predictions == 0) & (all_labels == 1))

print(f"Total Number of Testing Observations: {len(all_predictions)}")
print(f"True Positives: {TP}")
print(f"False Positives: {FP}")
print(f"True Negatives: {TN}")
print(f"False Negatives: {FN}")
```

## 6. Conclusion section: What is the conclusion of your 1st model? What can be done to possibly improve it?

We saw that our inital model appears to perform quite well. As stated above, to increase the performance, we will consider changing model architectural and runtime features for our next iteration. These improvements may include:


* **More epochs:** Training for more epochs would allow the model to continue learning, as the first couple of epochs will be needed just to reach a reasonable degree of accuracy, given the the model is randomly initialized. This is especially relevant because we chose to zero-pad non-square images so training for more epochs will enable the model to learn to ignore the zero-padding and to not use this feature when performing predictions.\
<br>
  
* **Adaptive learning rate:** Combined with a larger number of epochs especially, an adaptive learning rate (or at least one which changes throughout training a few times) may improve model performance. As the model gets increasingly accurate during epochs, a reduction in learning rate allows it to learn more slowly and not skip over a potential minimum.\
<br>
  
  
* **Increase training dataset size:** Right now we include 9600 images in our training dataset, and 1200 in our validation and test sets respectively, for a total of 12000 images across the three datasets. Because we have access to 60000 total images in our Kaggle dataset, we can increase the size of our train, validation, and test datasets to give the model more data to work with while training, and potentially increase it's ability to make more nuanced predicitions on unseen data.\
<br>
  
* **Randomize training data order:** We can randomize the order of the images in the training dataset, reshuffling between epochs, to avoid potential minor baises that may arise through the training data order.\
<br>
  
* **Model Architecture:** We plan on adding more convolutional layers to our model. Currently, we only have three convolutional layers that have feature maps of sizes 3, 32, and 64. We would like to add another convolutional layer for 128 feature maps so we can extract more abstract features from our images deeper in the model. We may also add model layers that consider brightness and sharpness of the images because in our exploratory data analysis, we observed variations in brightness and sharpness between AI generated and real images. We would like to distinguish these features further using model layers that address these features specifically.\
<br>

* **K-fold Cross Validation:** K-fold cross-validation helps improve our CNN model by providing a robust evaluation of its performance across different subsets of the data, ensuring it generalizes well to unseen images. By training on multiple folds and averaging results, it reduces the risk of overfitting or underfitting specific data splits, leading to more reliable model tuning and selection. \
<br>
    
* **Regularization:** We also plan to use regularization to help our model avoid overfitting. For example, we can use adversarial regularization to train the model not only with natural training data images but also perturbed data. By intentionally confusing the model with these perturbed images, we will encourage our model to become robust against adversarial perturbation and increase model performance.\
<br>
