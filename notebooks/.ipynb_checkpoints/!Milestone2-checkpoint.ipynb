<h3> EDA Step 3: Plotting Image Sizes across Classes </h3>

```python
#Specify types of file extensions

file_exts = [".jpg", ".png"]
real_widths = []
real_heights = []
fake_widths = []
fake_heights = []


# Confirm validity of given directory
if(not (os.path.exists(dir))):
    print("Error, specified directory does not exist")
    sys.exit(1)
```

```python
# Open all real images and store sizes
#Iterate through files in directory
for img_type in ["real", "fake"]:
    for file in os.listdir(os.path.join(dir, img_type)):
        filename = os.fsdecode(file)

        #Confirm file is an image
        for ext in file_exts:
            if(filename.endswith(ext)):

                #Open image
                img_path = os.path.join(dir, img_type, filename)
                with Image.open(img_path) as img:

                    #Save sizing to arrays
                    width, height = img.size
                    if(img_type == "real"):
                        real_widths.append(width)
                        real_heights.append(height)
                    else:
                        fake_widths.append(width)
                        fake_heights.append(height)
```

<h4> Sizing Analysis </h4>

For the consideration of image sizing and potential resizing need and strategy, this segment takes the filesize of all images and plots them to help us visualize the distribution of sizes. The included image set currently is only 30 (15 real, 15 fake) of the full set due to upload constraints, but when running this on 12,000 images locally (the entire testing set of the original dataset), there were some clear trends. We can see the distribution of real images has a much larger spread when compared to that of AI, and also seems to follow two common aspect ratios (indicated by the two trendlines we see in real images).

In terms of how we might consider mitigating this, we will likely be considering some form of resizing. This would need to occur regardless due to the Convolutional Neural Networks using fixed sized filters, which would likely make it difficult for the model to learn patterns; at the very least, varied sizes to actually train upon could introduce another dimension of error. However, knowing the original size may be a helpful feature, given the strong patterns we see above. 

In order to resize to a consistent standard, we will consider potentially cropping images or zero padding to fit aspect ratios, which could be a standard procedure of each image prior to being fed into the model. For some images, given that we have images as large as 12000x12000 or 10000x14000, we may need to resize down to a smaller size, such as 1024x1024 or even 512x512. We may also consider adding max pooling layers inside the CNN in order to resize the image and extract information within the CNN, rather than losing image data before the CNN.

Outliers purely based on size shouldn't prove to be too difficult given that we can resize them. However, we may experience aspect-ratio based outliers where the image proves challenging to resize to a specific aspect ratio, even when cosnidering cropping or zero padding options (if too much image data ends up getting lost). In that case we might just consider dropping those images, as changing the content of the primary part of the image could introduce unwanted error.

<h3> EDA Step 4: Plotting Aspect Ratios Across All Images </h3>

```python
# dimensions is (width, height) tuples for each image
dimensions = []

for img_path in images:
    img = cv2.imread(img_path)

    # Corrupt images identification
    if img is None:
        print("img_path: ", img_path)
        continue

    height, width = img.shape[:2]
    dimensions.append((width, height))  # Append as (height, width, # color channels) <-- we only grab the first 2


aspect_ratios = [width / height for width, height in dimensions]
plt.figure(figsize=(10, 6))
plt.hist(aspect_ratios, bins=20, color='skyblue', edgecolor='black')
plt.xlabel("Aspect Ratio (Width / Height)")
plt.ylabel("Frequency")
plt.title("Distribution of Aspect Ratios in the Dataset")
plt.show()
```

<h4> Aspect Ratio Analysis: </h4>

It seems that not all images are square or have an aspect ratio (width/height) of 1 as there are distinct peaks in the histogram between 0 and 2. We need all images to be the same size for the CNN input for training in batches because we need to apply the same convolutional and pooling operations uniformly across all images. Therefore, it is possible that cropping all images to 128 x 128 pixels could cause the images to loose important visual information required for the CNN layers to extract features. We may need to resize by adding padding for non square images to fit them into a square frame without losing aspect ratio integrity. Here are two possible solutions we propose and we plan on using one of these two solutions for our pre-processing based on the effectiveness and model accuracy for both. <br>

1 - Prioritizing aspect ratio: We will resize each image to a predefined square size of the (max image height and width) by (max image height and width) dimensions so all the image detailed are preseved. We will add padding to smaller dimension images to make them square and have the dimension (max image height and width) by (max image height and width). This way, we can preserve the aspect ration integrity for each image and prevent distortion of content. However, this may introduce empty areas via padding in images, which can add more noise for the CNN. 

2 - Not Prioritizing aspect ratio integrity: We will directly resize all images to a square shape of (128 by 128) without padding. This approach would sacrifice aspect ratio integrity, but will ensure consistent input size for the CNN. We will consider this approach if minor distortions are unlikely to impact the distingushing features of AI-generated vs. real images. This idea is simpler to implement and will prevent introducing empty spaces to images via padding, but may distort images and thereby impact model performance. 128 by 128 is a common size for images being passed into CNNs, which is why we chose that size. <br>

<h3> EDA Step 5: Plotting Image Brightness Across AI-Generated and Real Images </h3>

```python
ai_generated_brightness = []
real_brightness = []

for img_path, label in zip(images, labels): # iterate through image and class label
    # loading each image in grayscale mode to simplify brightness calculation by reducing to a single channel (0 to 255 pixel values)
    # grayscale represents intensity/brightness better because it gets the weighted average of the RGB color channels
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

    # Corrupt images identification
    if img is None:
        print("img_path: ", img_path)
        continue
    # average brightness: calculate the mean pixel intensity for the grayscale image, which represents the image's overall brightness
    average_brightness = np.mean(img)

    if label == 0:  # label 0 is for ai-generated images
        ai_generated_brightness.append(average_brightness)
    else:           # label 1 is for real images
        real_brightness.append(average_brightness)
```

```python
avg_real_brightness = np.mean(real_brightness)
avg_ai_generated_brightness = np.mean(ai_generated_brightness)

plt.figure(figsize=(6, 4))
plt.bar(['Real Images', 'AI-Generated Images'], [avg_real_brightness, avg_ai_generated_brightness], color=['blue', 'orange'])
plt.ylabel("Average Brightness")
plt.title("Average Brightness of Real vs. AI-Generated Images")
plt.show()
```

```python
# Generating indices for each image type
ai_generated_indices = list(range(len(ai_generated_brightness)))
real_indices = list(range(len(real_brightness)))

# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(ai_generated_indices, ai_generated_brightness, color='orange', alpha=0.5, label='AI-Generated Images')
plt.scatter(real_indices, real_brightness, color='blue', alpha=0.5, label='Real Images')
plt.xlabel("Image Index")
plt.ylabel("Brightness")
plt.title("Brightness Comparison between AI-Generated and Real Images")
plt.legend()
plt.show()
```

Here are some steps we plan to take:  <br>

2 - Minimize focus on Brightness: We want to ensure that brightness variations do not skew the model performance. We will normalize or standardize the pixel values by scaling the pixel intensity values of each image to a fixed range of 0-1 or doing Z score standardization. This brightness scaling to a standard range with keep the overall brightness levels of the images within a similar range, minimizing brightness differences without alterning other features. This can help reduce the influence of varying lighting conditions or overall brightness differences, allowing the model to focus on other distinguishing complex feautres since the difference in brightness is very minimal. 

3 - Maximize focus on brightness as distinguishing factor: We want to consider designing CNN layers with a focus on brightness gradients to see if this would contribute useful information to the CNN. <h3> EDA Step 6: Color Intensity Analysis </h3>

<h4> Plot Density Plot (All Pixels) </h4>

```python
def calculate_color_densities(image_paths):
    density_r = np.zeros(256)
    density_g = np.zeros(256)
    density_b = np.zeros(256)
    total_pixels = 0
    
    for path in image_paths:
        image = cv2.imread(path)
        if image is None:
            print(path)
            continue
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        total_pixels += image_rgb.shape[0] * image_rgb.shape[1]
        
        density_r += cv2.calcHist([image_rgb], [0], None, [256], [0, 256]).flatten()
        density_g += cv2.calcHist([image_rgb], [1], None, [256], [0, 256]).flatten()
        density_b += cv2.calcHist([image_rgb], [2], None, [256], [0, 256]).flatten()
    
    # Normalize to convert to density
    density_r /= total_pixels
    density_g /= total_pixels
    density_b /= total_pixels
    
    return density_r, density_g, density_b
```

```python
real_density_r, real_density_g, real_density_b = calculate_color_densities(real_images)
fake_density_r, fake_density_g, fake_density_b = calculate_color_densities(fake_images)
```

```python
# Plot the density histograms
plt.figure(figsize=(13, 3))

colors = {1 : "Red", 2 : "Green", 3: "Blue"}

for i in range(1, 4):
    plt.subplot(1, 3, i)
    color = colors[i]
    if color == "Red":
        plt.plot(real_density_r, color='red', label='Real')
        plt.plot(fake_density_r, color='darkred', label='Fake', linestyle='--')
    elif color == "Green":
        plt.plot(real_density_g, color='green', label='Real')
        plt.plot(fake_density_g, color='darkgreen', label='Fake', linestyle='--')
    else:
        plt.plot(real_density_b, color='blue', label='Real')
        plt.plot(fake_density_b, color='darkblue', label='Fake', linestyle='--')
    plt.title(color + ' Channel Density')
    plt.xlabel('Pixel Intensity')
    plt.ylabel('Density')
    plt.legend()

plt.tight_layout()
plt.show()
```

<h4> Plot Mean Color Value </h4>

```python
def calculate_mean_rgb(image_paths):
    mean_r = []
    mean_g = []
    mean_b = []
    
    for path in image_paths:
        image = cv2.imread(path)
        if image is None:
            print(path)
            continue
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        
        mean_r.append(np.mean(image_rgb[:, :, 0]))
        mean_g.append(np.mean(image_rgb[:, :, 1]))
        mean_b.append(np.mean(image_rgb[:, :, 2]))
    
    return mean_r, mean_g, mean_b
```

```python
real_mean_r, real_mean_g, real_mean_b = calculate_mean_rgb(real_images)
fake_mean_r, fake_mean_g, fake_mean_b = calculate_mean_rgb(fake_images)
```

```python
# Plot the density histograms
plt.figure(figsize=(13, 3))

colors = {1 : "Red", 2 : "Green", 3: "Blue"}

for i in range(1, 4):
    plt.subplot(1, 3, i)
    color = colors[i]
    if color == "Red":
        plt.plot(np.sort(real_mean_r), color='red', label='Real')
        plt.plot(np.sort(fake_mean_r), color='darkred', label='Fake', linestyle='--')
    elif color == "Green":
        plt.plot(np.sort(real_mean_g), color='green', label='Real')
        plt.plot(np.sort(fake_mean_g), color='darkgreen', label='Fake', linestyle='--')
    else:
        plt.plot(np.sort(real_mean_b), color='blue', label='Real')
        plt.plot(np.sort(fake_mean_b), color='darkblue', label='Fake', linestyle='--')
    plt.title("Mean " + color + ' Channel Density')
    plt.xlabel('Image')
    plt.ylabel('Pixel Intensity')
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
def calculate_median_rgb(image_paths):
    mean_r = []
    mean_g = []
    mean_b = []
    
    for path in image_paths:
        image = cv2.imread(path)
        if image is None:
            print(path)
            continue
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        
        mean_r.append(np.median(image_rgb[:, :, 0]))
        mean_g.append(np.median(image_rgb[:, :, 1]))
        mean_b.append(np.median(image_rgb[:, :, 2]))
    
    return mean_r, mean_g, mean_b
```

<h3> EDA Step 7: Image Sharpness Analysis </h3>

```python
ai_generated_sharpness = []
real_img_sharpness = []

for img_path, label in zip(images, labels):
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    if img is None:
        continue
    laplacian_var = cv2.Laplacian(img, cv2.CV_64F).var()
    if label == 0:
        ai_generated_sharpness.append(laplacian_var)
    else:
        real_img_sharpness.append(laplacian_var)
```

```python
real_img_data = pd.DataFrame(real_img_sharpness)
ai_generated_data = pd.DataFrame(ai_generated_sharpness)
print("real_img_data: ")
print(real_img_data.describe())

print("ai_generated_data: ")
print(ai_generated_data.describe())

avg_real_img_sharpness = np.mean(real_img_sharpness)
avg_ai_generated_sharpness = np.mean(ai_generated_sharpness)

```

<h4> Plotting image sharpness for both AI generated images and Real images </h4>

```python

# Plotting data points with mean lines for each category
plt.figure(figsize=(10, 6))

# Plot individual sharpness points
plt.plot(ai_generated_sharpness, label="AI Generated Sharpness", marker='o', linestyle='', markersize=6)
plt.plot(real_img_sharpness, label="Real Image Sharpness", marker='x', linestyle='', markersize=6)

# Adding mean lines for each category
plt.axhline(y=avg_ai_generated_sharpness, color='blue', linestyle='--', label="Avg AI Generated Sharpness")
plt.axhline(y=avg_real_img_sharpness, color='orange', linestyle='--', label="Avg Real Image Sharpness")

# Labels and Title
plt.xlabel("Image Index")
plt.ylabel("Sharpness (Variance of Laplacian)")
plt.title("Sharpness Comparison between AI Generated and Real Images with Averages")
plt.legend()

# Display the graph
plt.show()
```

<h4> Analysis </h4>

To analyze image sharpness quantitatively, we utilized the Laplacian method in OpenCV to emphasize areas with rapid intensity changes, which often correspond to edges and details in the image.


Laplacian Computation: Using cv2.Laplacian(img, cv2.CV_64F), we computed the Laplacian of each image, highlighting areas with rapid intensity changes (i.e., edges and details). 

Variance of Laplacian Values: Calculating the variance of the Laplacian values (via .var()) serves as an effective measure of sharpness.

Higher variance suggests a sharper image, as more edges and details are present.
Lower variance indicates a blurrier image with fewer intense changes in pixel values, resulting in a smoother appearance.

1. Sharper Images: AI-generated images tend to have higher sharpness on average (indicated by a higher mean sharpness). This could imply that AI models are tuned to produce images that appear more defined or crisp.

2. Less Variation in AI Sharpness: The lower standard deviation in AI-generated images implies that the sharpness is more consistent across the dataset, with fewer variations in sharpness levels compared to real images.

3. Higher Median Sharpness: The higher median in AI-generated images suggests that even the "middle" sharpness values in AI images are higher than those in real images, reinforcing the consistency in sharpness across AI images.

This analysis suggests AI images are optimized for sharper, more consistent quality, whereas real images display a broader range of sharpness, likely due to diverse conditions in real-life photography (lighting, focus, etc.).
