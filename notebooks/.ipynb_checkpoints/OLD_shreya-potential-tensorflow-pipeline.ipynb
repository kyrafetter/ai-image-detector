```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# Set image parameters
img_height, img_width = 128, 128
batch_size = 32

# Define ImageDataGenerator for training and validation
train_datagen = ImageDataGenerator(
    rescale=1.0 / 255,           # Normalize pixel values
    validation_split=0.2          # Split for validation data
)

# Load the train and validation sets
train_data = train_datagen.flow_from_directory(
    '/Users/shrey/ai-image-detector/data/images',          # Directory path
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='binary',          # For binary classification
    subset='training'
)
```

```python
train_data
```

```python
import matplotlib.pyplot as plt

# Retrieve one batch of images and labels from train_data
# batch size is 32
images, labels = next(train_data)

# Display a few images from the batch
plt.figure(figsize=(10, 10))
for i in range(9):  # Display 9 images (3x3 grid)
    plt.subplot(3, 3, i + 1)
    plt.imshow(images[i])
    plt.title(f"Label: {int(labels[i])}")  # Convert label to int for binary classification
    plt.axis("off")
plt.show()
```
