```python
import matplotlib.pyplot as plt
from PIL import Image
import os
import sys


#Specify location of images and types of file extensions
# dir should be the directory containing "real" and "fake" directories of images
dir = "../../data/images"
file_exts = [".jpg", ".png"]
real_widths = []
real_heights = []
fake_widths = []
fake_heights = []


# Confirm validity of given directory
if(not (os.path.exists(dir))):
    print("Error, specified directory does not exist")
    sys.exit(1)
```

```python
# Open all real images and store sizes
#Iterate through files in directory
for img_type in ["real", "fake"]:
    for file in os.listdir(os.path.join(dir, img_type)):
        filename = os.fsdecode(file)

        #Confirm file is an image
        for ext in file_exts:
            if(filename.endswith(ext)):

                #Open image
                img_path = os.path.join(dir, img_type, filename)
                with Image.open(img_path) as img:

                    #Save sizing to arrays
                    width, height = img.size
                    if(img_type == "real"):
                        real_widths.append(width)
                        real_heights.append(height)
                    else:
                        fake_widths.append(width)
                        fake_heights.append(height)
```

```python
#Plot sizes as coordinate points of (width, height)
plt.plot(real_widths,real_heights, 'ro', fake_widths,fake_heights, 'bo')
plt.xlabel('Image Width')
plt.ylabel('Image Height')
plt.legend(['Real Images', 'Fake Images'])
plt.show()
```

For the consideration of image sizing and potential resizing need and strategy, this segment takes the filesize of all images and plots them to help us visualize the distribution of sizes. The included image set currently is only 30 (15 real, 15 fake) of the full set due to upload constraints, but when running this on 12,000 images locally (the entire testing set of the original dataset), there were some clear trends. We can see the distribution of real images has a much larger spread when compared to that of AI, and also seems to follow two common aspect ratios (indicated by the two trendlines we see in real images).

In terms of how we might consider mitigating this, we will likely be considering some form of resizing. This would need to occur regardless due to the Convolutional Neural Networks using fixed sized filters, which would likely make it difficult for the model to learn patterns; at the very least, varied sizes to actually train upon could introduce another dimension of error. However, knowing the original size may be a helpful feature, given the strong patterns we see above. In order to resize, we will consider potentially cropping images or zero padding to fit aspect ratios, which could be a standard procedure of each image prior to being fed into the model.

Outliers purely based on size shouldn't prove to be too difficult given that we can resize them. However, we may experience aspect-ratio based outliers where the image proves challenging to resize to a specific aspect ratio, even when cosnidering cropping or zero padding options (if too much image data ends up getting lost). In that case we might just consider dropping those images, as changing the content of the primary part of the image could introduce unwanted error.
